# Question 1

A simple linear regression is a statistical model where a linear line is used to approximate a relationship between an input predictor variable and output outcome variable.


The outcome variable is the output and dependent variable of the statistical problem. Typically, its the variable that we are using simple linear regression to accurately predict with a linear line. It is represented on the y-axis and must be numerical. 

The predictor variable is the input and independent variable of the statistical problem. Typically, its the variable that determines the outcome variable. The predictor variable can be either numerical or categorical. It is represented in the x-axis.  

- The SLOPE COEFFICIENT represents the linear relationship between the outcome variable and the predictor variable. It represents the rate of change in the outcome (y) variable with an increase in the predictor (x) variable. 

- The INTERCEPT COEFFICIENT is effectively the y-intercept when x is 0. In the context of simple linear regression, it is the value of the outcome variable when the predictor variable is 0. 

- The ERROR TERM is an underlying assumption that the linear relationship is not perfect, and that variation will exist in the data when accounting for the distance to the line (which is the residual). The squared residuals from the variation of the data is used to calculate R^2

- R^2 is the amount variation of the outcome variable that is explained by the predictor variable. In simple terms, a high value of R^2 demonstrates a strong relationship between the predictor and outcome variables.

*For clarification; a simple linear regression only has 1 predictor variable. Other forms of linear regression, such as multiple linear regression, has multiple predictor variables*

#### To generate a sample from the sample linear regression model
- You must define the predictor value and have a set of values for it. 
- The Slope and intercept coefficients must be chosen 
- the error term must be identified and calculated through variance or standard deviation
- Afterwards, the outcome variable can be calculated with the linear regression model

These following assumptions must always exist for the linear regression model:
- the output and predictor variables must have a linear relationship
- the error term (standard deviation) should have a normal distribution (e.g standard deviation of 1)
- Homoscedasticity: the variance of the data is relatively uniform and equal
- Independence: the data is independent from each other

Contextually, if we were to conduct multiple scenarios of random sampling for the theoretical simple linear regression model, the error term would result in different sets of data points being generated in each scenario. However this would also result in a different fitted linear regression line for each set (i.e a different slope and intercept). This can be seen below in the graph that is coded. 

Regardless, there is a true intercept and slope for the fitted linear regression line (vertical red line on the graph below) that Code 2 cannot compute objectively as a result of the error term effecting the generated data points from Code 1.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.graph_objects as go

# Define the theoretical simple linear regression model parameters
true_slope = 2.5
true_intercept = 5
sigma = 2
num_samples = 100  # Number of samples to generate
sample_size = 100  # Number of data points per sample

# Arrays to store the estimated slopes and intercepts
estimated_slopes = []
estimated_intercepts = []

# Generate multiple samples and fit the model to each sample
np.random.seed(0)
for _ in range(num_samples):
    x = np.random.rand(sample_size) * 10  # Generate predictor variable (X)
    true_regression_line = true_slope * x + true_intercept
    y = true_regression_line + np.random.randn(sample_size) * sigma  # Simulate points with noise

    # Create a DataFrame from the simulated data
    df = pd.DataFrame({'x': x, 'Y': y})

    # Fit a Simple Linear Regression model using statsmodels
    model = smf.ols("Y ~ x", data=df).fit()

    # Store the estimated coefficients
    estimated_slopes.append(model.params['x'])
    estimated_intercepts.append(model.params['Intercept'])

# Convert lists to arrays for easier analysis
estimated_slopes = np.array(estimated_slopes)
estimated_intercepts = np.array(estimated_intercepts)

# Create histogram for estimated slopes using Plotly
fig_slope = go.Figure()

fig_slope.add_trace(go.Histogram(
    x=estimated_slopes,
    nbinsx=20,
    name='Estimated Slopes',
    marker_color='blue',
    opacity=0.75
))

# Add a vertical line for the true slope
fig_slope.add_shape(
    type="line",
    x0=true_slope,
    y0=0,
    x1=true_slope,
    y1=20,
    line=dict(color="red", width=2, dash="dash"),
    name='True Slope'
)

# Update layout for slope histogram
fig_slope.update_layout(
    title_text='Distribution of Estimated Slopes',
    xaxis_title_text='Estimated Slope',
    yaxis_title_text='Frequency'
)

# Create histogram for estimated intercepts using Plotly
fig_intercept = go.Figure()

fig_intercept.add_trace(go.Histogram(
    x=estimated_intercepts,
    nbinsx=20,
    name='Estimated Intercepts',
    marker_color='green',
    opacity=0.75
))

# Add a vertical line for the true intercept
fig_intercept.add_shape(
    type="line",
    x0=true_intercept,
    y0=0,
    x1=true_intercept,
    y1=20,
    line=dict(color="red", width=2, dash="dash"),
    name='True Intercept'
)

# Update layout for intercept histogram
fig_intercept.update_layout(
    title_text='Distribution of Estimated Intercepts',
    xaxis_title_text='Estimated Intercept',
    yaxis_title_text='Frequency'
)

# Show the plots with specified renderer
fig_slope.show(renderer="png")
fig_intercept.show(renderer="png")
```

# Question 5 

The "ordinary least squares" method in simple linear regression is where you choose a line that has the smallest possible value of squared residuals (for each data point in the model) as your line for the linear regression model. 

From a theoretical standpoint, the residuals in this method are intentionally squared for two reasons:
- It removes negative values (residuals under the line) that would skew the analysis of the line (for example, the possible case of negative and positive values cancelling each other out and making the line look accurate; not an accurate analysis)
- It makes large residuals easier to analyze. For example a residual of 2 and 10 seems somewhat small, however when squared, it becomes 4 and 100. The squared scenario is easier to interpret than the non-squared scenario.

   - Simple linear regression involves modeling the relationship between a single predictor \( x \) and an outcome \( Y \).

```python
import plotly.express as px
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define the long wait limit
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Print the summary of the fitted model for long wait times
print(smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1])

# Create a scatter plot with a linear regression trendline
fig = px.scatter(old_faithful[long_wait], x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions for long wait times (>" + str(long_wait_limit) + ")", 
                 trendline='ols')
fig.show(renderer="png")

# Bootstrap sampling to collect and visualize the bootstrapped sampling distribution of the fitted slope coefficients
n_bootstrap = 1000
bootstrap_slopes = []

for _ in range(n_bootstrap):
    bootstrap_sample = old_faithful.sample(n=len(old_faithful), replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Plot the bootstrapped sampling distribution of the fitted slope coefficients
plt.hist(bootstrap_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Bootstrapped Sampling Distribution of Slope Coefficients')
plt.show()

# Simulate samples under the null hypothesis of no linear association
n_simulations = 1000
simulated_slopes = []
n = 160

for _ in range(n_simulations):
    simulated_duration = np.random.normal(loc=old_faithful['duration'].mean(), scale=old_faithful['duration'].std(), size=n)
    simulated_data = old_faithful[['waiting']].sample(n=n, replace=True).copy()
    simulated_data['duration'] = simulated_duration
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slopes.append(model.params['waiting'])

# Plot the sampling distribution of the fitted slope coefficients under the null hypothesis
plt.hist(simulated_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of Slope Coefficients under Null Hypothesis')
plt.show()

# Calculate the 95% bootstrapped confidence interval for the slope coefficient
ci_lower = np.percentile(bootstrap_slopes, 2.5)
ci_upper = np.percentile(bootstrap_slopes, 97.5)
print(f"95% Bootstrapped Confidence Interval for Slope Coefficient: ({ci_lower}, {ci_upper})")

# Check if 0 is contained within the 95% bootstrapped confidence interval
contains_zero = ci_lower <= 0 <= ci_upper
print(f"Does the 95% bootstrapped confidence interval contain 0? {contains_zero}")

# Calculate the p-value from the simulated slopes under the null hypothesis
p_value_simulated = np.mean(np.array(simulated_slopes) >= smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params['waiting'])
print(f"Simulated p-value: {p_value_simulated}")
```
